Early use of continuous positive airway pressure (CPAP) and early selective use of surfactant
in preterm infants with respiratory distress syndrome (RDS) is recommended by
international guidelines 1 2 . This recommendation is based on systematic reviews which have
shown that compared to the ‘’intubation group’’, infants receiving CPAP at birth had
reduced death and /or bronchopulmonary dysplasia (BPD) 3 . A Cochrane review has shown
that prophylactic use of surfactant is associated with a higher risk of death /BPD 4 , and the
review suggested early selective use of surfactant rather than prophylactic surfactant use 5 .
Early selective use of surfactant has been shown to have significant reduction in neonatal
morbidities 6 . Criteria for the selection of preterm infants for surfactant administration vary
between neonatal units, agencies and countries 7 . In clinical practice, varying levels of
fraction of inspired oxygen (FiO2), along with other clinical factors, are used to select infants
for surfactant administration. The reliance on FiO 2 thresholds lacks strong evidence 8 9 .
Infants could take variable time to reach this threshold, which could reduce the benefits of
early surfactant administration 10 .
Emerging evidence suggests that lung ultrasound (LU) could predict the need for surfactant
early and BPD in the preterm population 11-14 . LU can be performed easily at the bedside
requires a very short period (3mins ± 2 12 ), is non-invasive, no radiation hazard (unlike X-
rays), has a steep learning curve and has a good inter-rater reliability (k=0.89) 11 12 .
The primary objective of our study was to assess the diagnostic value of LUS performed by
operators of various levels of experience and identify the best lung ultrasound score (LUS),
using ten versus six zones, before 3hrs of age to predict the need for surfactant in preterm
infants ≤ 34 weeks on non-invasive respiratory support. The secondary objectives of our
study were as follows: (1) to study the value of serial LU until Day 3 and to predict the need
for repeated doses of surfactant, (2) to predict the development of BPD at 36 weeks, and (3)
to report the correlation between chest X-ray and LUS, and between Silverman-Anderson
score and LUS.

Methods:
We conducted a multi-centre observational prospective study in three level 3 neonatal
intensive care units in the United Kingdom (Centre 1: James Cook University Hospital,
Middlesbrough (JCUH); Centre 2: University Hospital Wishaw, Glasgow (UHW) and Centre 3:

5
Oliver Fisher Neonatal unit, Kent (OFNU). We recruited inborn preterm infants with a birth
gestational age of ≤34 weeks and receiving non-invasive respiratory support. We excluded
infants receiving surfactant or delivery room intubation before the first LU, infants with
major congenital malformations, and infants with pneumothorax needing intervention.
For chest X-rays, we used scoring developed by Perri et al. 15 and Silverman-Anderson score
for assessment of respiratory distress 16 . We defined BPD as oxygen or respiratory support
needed at 36 weeks postmenstrual age for preterm infants &lt; 32 weeks birth gestational
age 17 .

LU training: We have organised one full-day LU workshop independently at each of the
centres. Following this, we also developed an online learning module. All scans were
performed by the local team members who had completed one or more of the
aforementioned formal LU training sessions. Operators were divided into four experience
levels: “Training only” (completed only online course and workshop); “Beginner” (performed
20 scans with 5-10 supervised scans); “Intermediate” (50 to 70 LU scans of all pathologies)
and “Advanced” (more than 70 LU scans of all pathologies). For the operators “Training
only” and “Beginner”, we suggested performing at least the first five LU scans under
supervision.

Lung ultrasound: All of the recruited infants underwent LU before surfactant administration
within 3 hours of life and subsequent serial LU every 12-24hrs until day 3 (total of 4 LU). We
chose 3 hours cut-off as it aligns with the preferred timing for early surfactant
administration 5 . For each of the scans, we assessed five lung zones on each side (A total of
10 lung zones): Upper Anterior at the level of the mid-clavicular line, Lower Anterior at the
level of the mid-clavicular line, Lateral at the level of the mid axillary line, Upper Posterior at
the level of posterior axillary line between scapula and spine and Lower posterior at the
same level. For infants in the supine position at the time of LU, we tilted the infant to each
side briefly to obtain LU images for the posterior region and vice versa for infants who are
prone at the time of LU. Tilting was performed based on the clinical stability of the infant.
We used the LU scoring system adapted from Brat et al. 11 and assessed each area of interest
(5 areas for each lung: 10 lung zones) for a score of 0 to 3 for each zone, based on the four
different patterns.

6
Each unit had its guidelines for respiratory management, and the decision to intubate was
under the clinical discretion of the attending team 2 . For the recruited infants, members of
the clinical team who underwent training performed LU and calculated their scores. They
also completed the case report form (workbook) for each of the LU (Appendix_1). The LU
findings were not shared with the clinical team (except in cases of pneumothorax), who
pragmatically treated those babies according to their local guidelines. We collected infant
and maternal demographics and clinical data from the neonatal database. Only non-
identifying study data were entered and managed using REDCap (Research Electronic Data
Capture), a secure online research database 18 by the research team. Video files of the LU
were downloaded without any patient information and securely transferred to one of the
expert study investigators (VMP), who was blinded to the clinical history and LUS calculated
by the person who performed the LU at the bedside. We used the scores of the expert
(VMP) for all our primary analyses. An expert (VMP) labelled a LU as “uninterpretable” if the
quality of the scans was too poor (e.g., too dark) and, as a result, unable to score. Details of
the study protocol are provided in Appendix_2.
Descriptive statistics were used for population characteristics. Categorical variables
were presented as proportions, while numerical variables were presented as mean with
standard deviation (SD) or median with interquartile range (IQR) as appropriate. We
calculated sensitivity, specificity, and predictive value for LUS for the need for surfactant and
BPD at 36 weeks. Receiver operator characteristic (ROC) curves were constructed, and the
corresponding areas under the curve (AUCs) were reported. Comparisons between AUCs
were performed using the DeLong method for ROC curves. We specifically reported
diagnostic values for the LUS cut-off of 8 and 9 as these scores were commonly reported to
have higher diagnostic values in previous studies 11 13 .
We used Cohen&#39;s kappa (κ) as a measure of inter-rater agreement for categorical
scales between the expert (VMP) and other operators LUS 19 . For assessing the correlations
between two sets of continuous variables, we used Spearman’s rank correlation coefficient
and the strengths of correlation were reported as described by Schober et al 20 . For all
statistical analyses, a p-value of &lt;0.05 was considered significant. All the analyses were
performed using StataNow/SE 18.5 (StataCorp LLC). We used Strengthening the Reporting of

7
Observational Studies in Epidemiology (STROBE) statement guidelines for reporting the
study (Appendix 3) 21 .
Currently, chest X-ray is the standard diagnostic tool for imaging infants with
respiratory distress. The sensitivity of chest X-ray in predicting the need for surfactant was
38% 13 , to improve the sensitivity to approximately 60% by LU, we needed 39 infants to show
this difference with a power of 90% and alpha error of 5% 22 . With the expected loss of poor-
quality images, we aimed to recruit at least 50 infants with at least half of the population
(25 babies) in the gestation ≤29 weeks. We obtained retrospective consent from the
mothers of the recruited infant. This study was registered with clinictrials.org
(NCT05782569), and ethics board approval was obtained (IRAS 291372).

```python
#which are the most predictive regions for surfactant =0
import pandas as pd
from sklearn.metrics import confusion_matrix

# Assuming your DataFrame is named df1

for region in ['R1', 'R2', 'R3', 'L1', 'L2', 'L3']:
    # Create a binary prediction based on region = 1 or 1
    y_pred = (df1[region] != 2 ).astype(int)

    # Get the true labels for Surfactant
    y_true = df1['Surfactant'] == 0

    # Calculate the confusion matrix
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()

    # Calculate sensitivity and specificity
    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)

    print(f"Region: {region}")
    print(f"Sensitivity: {sensitivity:.2f}")
    print(f"Specificity: {specificity:.2f}")
    print( "Youden's index=", sensitivity + specificity - 1)
    print("-" * 20)  # Print a separator line
```

```python
# Combination of 2 regions with score=<1
import pandas as pd
from sklearn.metrics import confusion_matrix
from itertools import combinations

# Assuming your DataFrame is named df1

regions = ['R1', 'R2', 'R3', 'L1', 'L2', 'L3']

# Get true labels for no surfactant need (Surfactant = 0)
y_true = (df1['Surfactant'] == 0).astype(int)

# Iterate through all combinations of two regions
for region_pair in combinations(regions, 2):
    region1, region2 = region_pair

    # Create a binary prediction based on both regions <= 1
    y_pred = ((df1[region1] <= 1) & (df1[region2] <= 1)).astype(int)

    # Calculate the confusion matrix
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()

    # Calculate sensitivity and specificity
    sensitivity = tp / (tp + fn) if (tp + fn) != 0 else 1
    specificity = tn / (tn + fp) if (tn + fp) != 0 else 1

    # Youden's J statistic (for overall performance)
    youden_j = sensitivity + specificity - 1

    print(f"Regions: {region1}, {region2}")
    print(f"Sensitivity: {sensitivity:.2f}")
    print(f"Specificity: {specificity:.2f}")
    print(f"Youden's J: {youden_j:.2f}")
    print("-" * 20)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt
import numpy as np

# Assuming your DataFrame is named df1

# Select features and target
features = ['R1', 'R2', 'R3', 'R4', 'R5', 'L1', 'L2', 'L3', 'L4', 'L5', 'GA']
target = 'Surfactant'

X = df1[features].copy()
y = df1[target]

# Ensure features are integers (1 or 2)
for feature in features:
    X[feature] = pd.to_numeric(X[feature], errors='coerce')
    X[feature] = X[feature].apply(lambda val: 1 if val < 1.5 else 2)

# Pre-classify instances with L3 <= 1 and R3 <= 1
mask = (X['L3'] == 1) & (X['R3'] == 1)
X['modified_target'] = y.copy()
X.loc[mask, 'modified_target'] = 0
y_modified = X['modified_target']

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X[features], y_modified, test_size=0.2, random_state=42)

# Train the decision tree
tree_classifier = DecisionTreeClassifier(random_state=42, max_depth=3)
tree_classifier.fit(X_train, y_train)

# --- Enforce L3 as the first split and remove it from further splits ---
L3_index = features.index('L3')
tree_classifier.tree_.feature[0] = L3_index
tree_classifier.tree_.threshold[0] = 1.5

# Remove L3 from the feature list for subsequent splits
remaining_features = features[:L3_index] + features[L3_index + 1:] # keep all except L3

# --- Re-train the tree with the remaining features for the sub-trees---
# Create a mask to identify the samples that go to the left and right sub-trees:
# Left subtree: L3 <= 1
# Right subtree: L3 > 1

mask_left = (X_train['L3'] <= 1)
mask_right = (X_train['L3'] > 1)

# Extract the indices of the samples for each subtree
left_indices = X_train[mask_left].index
right_indices = X_train[mask_right].index

# Retrain the left subtree
if len(left_indices) > 0: # if there are samples going to the left subtree
    left_subtree = DecisionTreeClassifier(random_state=42, max_depth=2) # max_depth is reduced by 1
    left_subtree.fit(X_train.loc[left_indices, remaining_features], y_train.loc[left_indices])

# Retrain the right subtree
if len(right_indices) > 0: # if there are samples going to the right subtree
    right_subtree = DecisionTreeClassifier(random_state=42, max_depth=2) # max_depth is reduced by 1
    right_subtree.fit(X_train.loc[right_indices, remaining_features], y_train.loc[right_indices])

# --- End of split enforcement ---


# Modify thresholds for visualization
thresholds = tree_classifier.tree_.threshold
for i in range(len(thresholds)):
    if thresholds[i] == 1.5:
        thresholds[i] = 1

# Predictions and evaluation
y_pred = tree_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print(report)

# Visualize the decision tree
plt.figure(figsize=(12, 8))
plot_tree(tree_classifier,
          feature_names=features,
          class_names=['No Surfactant', 'Surfactant'],
          filled=True,
          impurity=False)
plt.show()
```
